```python
import pandas as pd
import numpy as np
```

```python
xl = pd.read_excel("SampleData.xlsx", sheet_name="SalesOrders")
```

```python
xl.head()
```

```python
# prompt: Using dataframe xl: Pie Chart for total Sales(Total column) per Region(Region column)

import altair as alt

# Calculate total sales per region
region_sales = xl.groupby('Region')['Total'].sum().reset_index()

# Create the pie chart
alt.Chart(region_sales).mark_arc().encode(
    theta='Total',
    color='Region',
    tooltip=['Region', 'Total']
).properties(title='Total Sales per Region')
```

```python
# prompt: total sales per sales representative(Rep column)

import pandas as pd
import altair as alt

# Assuming xl DataFrame is already loaded as in the previous code

# Calculate total sales per sales representative
rep_sales = xl.groupby('Rep')['Total'].sum().reset_index()

# Create the bar chart
alt.Chart(rep_sales).mark_bar().encode(
    x='Rep',
    y='Total',
    tooltip=['Rep', 'Total']
).properties(title='Total Sales per Sales Representative')
```
